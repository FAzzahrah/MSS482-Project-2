Loan default and combined income play crucial rule in measure loan amount. Combined income provides a measure of the borrower's overall financial capacity by considering the income of both the applicant and the coapplicant. By including combined income as an independent varaible in the model, it allows for an assessment of the applicant's ability to afford a higher loan amount. Higher combined income suggest a greater capacity to handle larger loan amounts.

Meanwhile, loan default can provide insights into the applicant's creditwothiness and past repayment behavior. By including loan default in this analysis, the model can assess the relations between loan default and loan amount. This can help evaluate the risk associated with higher loan amounts, as applicants with a history of default may be more likely to default on larger loans.

### 3.1.8 T-test

```python
statistic, p_value = stats.ttest_ind(applicant['CombinedIncome'][applicant['Loan_Default'] == 'N'],
                applicant['CombinedIncome'][applicant['Loan_Default'] == 'Y'])


print(f"t's test statistic: {statistic}")
print(f"P-value: {p_value}")

alpha=0.05

if p_value > alpha:
    print("The means of the two groups are approximately equal.")
else:
    print("The means of the two groups are significantly different.")
```

### 3.1.10 Levene's test

```python
stat, p_levene = levene(scale(ld_n['CombinedIncome']), scale(ld_y['CombinedIncome']))

print(f"Homogeneity of Variance Statistic: {stat, p_levene}")


alpha = 0.05  # Significance level

if p_levene > alpha:
    print("Sample looks like it came from a normal distribution (fail to reject H0)")
else:
    print("Sample does not look like it came from a normal distribution (reject H0)")
```

---

### 3.1.11 Shapiro-Wilk test

```python
stat1, p_value1 = shapiro(scale(ld_n['CombinedIncome']))
stat2, p_value2 = shapiro(scale(ld_y['CombinedIncome']))

print(f'Shapiro-Wilk test for Loan Default (N): Statistic={stat1}, p-value={p_value1}')

alpha = 0.05  # Significance level

if p_value1 > alpha:
    print("Sample looks like it came from a normal distribution (fail to reject H0)")
else:
    print("Sample does not look like it came from a normal distribution (reject H0)")


print(f'\nShapiro-Wilk test for Loan Default (Y): Statistic={stat2}, p-value={p_value2}')

alpha = 0.05  # Significance level

if p_value2 > alpha:
    print("Sample looks like it came from a normal distribution (fail to reject H0)")
else:
    print("Sample does not look like it came from a normal distribution (reject H0)")
```

### 3.1.14 Linear Regression on Loan Default N

```python
X = sm.add_constant(ld_n['CombinedIncome'])
y = ld_n['LoanAmount']

model = sm.OLS(y, X).fit()
print(model.summary())

# Plotting the data and regression line
plt.scatter(X['CombinedIncome'], y)
plt.plot(X['CombinedIncome'], model.predict(X), color='black', linewidth=3)
plt.xlabel('Combined Income')
plt.ylabel('Loan Amount')
plt.title('Linear Regression with Statsmodels')
plt.show()
```

### 3.1.15 Linear Regression on Loan Default Y

Levene's test and Shapiro-Wilk test are statistical methods utilized to assess the assumptions of the t-tests. Levene's test assesses the homogeneity of variances assumption, which states that the variances of the populations being compared are equal meanwhile the Shapiro-Wilk test assesses the normality assumption, which states that the data within each group or sample is drawn from a normally distributed population. Levene's test helps verify the assumption of homogeneity of variances, while the Shapiro-Wilk test helps verify the assumption of normality. The Homogeneity of variance statistic obtained from Levene's test is (0.1325179295105228, 0.7166040738757835) where the test statistic is 0.132517929510522 and the p-value associated with the test is 0.7166040738757835. Once more, the p-value is greater than 0.05, leading to the same conclusion that the null hypothesis cannot be rejected. Based on the Shapiro-Wilk test, the samples exhibit characteristics of a normal distribution, as evidenced by the failure to reject the null hypothesis. The Shapiro-Wilk test results for both groups indicate that the assumption of normality is met. This allows for the implementation of a normality histogram.

### 3.2.2 INTERPRETATION ON OLS REGRESSION TABLE

The most significant metric that this summary OLS Regression results generate is R-squared. R-squared is the assessment of how much of the independent variable ("CombinedIncome" of "LoanDefault No") is explained by changes in our dependent variables. With a percentage of 0.602, our model accounts for 60.2% of the variation in our "LoanAmount" variable which indicates a moderate level of goodness of fit. Our adjusted R-Squares, 0.594 slightly lower than the R-Squared value itself which illustrates how adding additional variable is penalised. The obtained F-Statistic is 74.08 and Probability (Prob F-Statistic) is 2.29e-11 which quites low. According to the very low p-vaue of "CombinedIncome" (0.000), the F-statistic of 70.48 considered high in the sense that the evidence against the null hypothesis is strong. This inferred that the model is statistically significant. Because the p-values for each coefficient test the null hypothesis that the corresponding coefficient is equal to zero, the low p-value for "CombinedIncome" (0.000) in this instance indicates that it is a significant predictor. The coefficients of both constant(intercept) and "CombinedIncome" are 12.5131 and 0.0220. These are the coefficients for the regression equation which in this model, the regression equation is LoanAmount = 12.5131 + 0.0220*CombinedIncome.The 95% confidence interval for the variable CombinedIncome in this instance is [0.017, 0.027]. The summary table also shows the diagnostic statistics that provide insights into various aspects of the regression model and its residuals. The residuals' normality is evaluated using the Omnibus test. A test statistic of 3.440 indicates a deviation from normalcy, a low p-value (usually less than 0.05) would suggest that the residuals deviate considerably from a normal distribution. The Jarque Bera test also assessed for the normalcy residual which in our case the obtained value is 2.783 with given low p-value suggest that a deviation from the norm.
